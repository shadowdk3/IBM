## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
import requests
from bs4 import BeautifulSoup as bs
```

Download the webpage at the url
response = requests.get(url)

```python
#your code goes here
response = requests.get(url)
```

Create a soup object


```python
#your code goes here
soup = bs(response.content,"html.parser")
print(soup)
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
rows = soup.find_all('tr')[1:]
popular_language =[]

# Extract Language name and Annual Average Salary from each row
for row in rows:
    language = row.find_all('td')[1].get_text()
    salary = row.find_all('td')[3].get_text()
    print(f"Language: {language}, Salary: {salary}")
    popular_lan=[language,salary]
    popular_language.append(popular_lan)

print(popular_language)
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here

# your code goes here
import csv
with open('popular-languages.csv','w',newline='') as file:
    csvwriter = csv.writer(file)
    for row in popular_language:
        csvwriter.writerow(row)
        
import pandas as pd
df = pd.read_csv('popular-languages.csv')
df.head(50)

import matplotlib.pyplot as plt
import numpy as np

perfomance = [130801, 114383,113865,110981,101013,94082,92037,88726,84727,84793]
dfs = ['Swift','Python','JAVA','C++','JavaScript','GO','R','C#','SQL','PHP']
tks= np.arange(len(dfs))
plt.figure(figsize=(16,7))
plt.bar(tks, perfomance, align='center',alpha=0.5)
plt.xticks(tks,dfs)
plt.title('Programming Language')
plt.show()
```
